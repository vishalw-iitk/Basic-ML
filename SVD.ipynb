### SVD manually

```python
A_tr = A.transpose()
```

```python
Z1 = np.matmul(A, A_tr) #Covariance matrix w.r.t rows
Z2 = np.matmul(A_tr, A) #Covariance matrix w.r.t columns
```

```python
eigenvalues_z1, eigenvectors_z1 = np.linalg.eig(Z1)
eigenvalues_z2, eigenvectors_z2 = np.linalg.eig(Z2)
```

```python
U = eigenvectors_z1
V = eigenvectors_z2
```

```python
V.shape
```

```python
V_tr = V.transpose()
```

```python
# PCA level transformation => T(100, 30) matmul(A(100, 30), V(30, 30))
T1 = np.matmul(A, V)
T1.shape
```

```python
# A = U * Sigma * V_tr
# A(100,30) = U(100,100) * Sigma(100, 30) * V_tr(30, 30)
```

```python
import math
Sigma = np.zeros((100,30))
incr = 0
for j in range(Sigma.shape[1]):
    Sigma[j][j] = math.sqrt(eigenvalues_z1[incr])
    incr = incr + 1
# Sigma
```

```python
# Another method for projection
# T(100, 30) = matmul( U(100, 100) , Sigma(100, 30) )
T2 = np.matmul(U, Sigma)
T2.shape
```
